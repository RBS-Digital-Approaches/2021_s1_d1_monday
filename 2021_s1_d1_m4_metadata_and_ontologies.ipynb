# I.  Introduction

As digital Bibliographers we work both with digitized versions of original artifacts, and, digitally delivered metadata about artificats.  Metadata, a set of data that describes and gives information about an object or other data, is not a digital concepts, and has been used to help organize and facilitate searh and discovery of reasources since long before the first computers.  The original library card catalog was a complext, analog, metadata system.  

Today, most library and museum catalogs collect, store, and deliver metadata digitally.  And the digital catalogs of cultural heritage institutions provide rich data sources for those who wish to perform larg-scae analysis of cultural artifacts.  

But it is important realize that, like all data, metadata is not unbiased.  Both the raw, descriptive data itself and the mechanisms of storage and delivery introduce particular perspectives that faciliate certain types of scholarly quesions and preclude others.  This unit introduces the metadata strategies most widely used by cultural heritige institution with a particular eye towards their unique advantages and limitations.

# II. General Types of Metadata

All metadata is instantiated as one of two, primary types:

**1.  Unstructured Data:**  Unstructured data present information about objects in a non-normalized, unstructured manner.  For example, a set of reader's, free text notes about items in a collection is unstructured data.  Unstructured data can contain a tremendous amount of information about an artifact, and particularly flexible in it ability to present nuanced information.  But it is only minimally useful (without a lot of pre-processing, segmentation, and extraction) for comparitive, computational analysis.  As such, digital metadata systmes tend to favor structured representations of data, as described below.

**2.  Structured Data:**  Structured data imposes, as the name implies, a structure on the metatada being captured and represented.  Structured data systems establishes pre-defined categories of information to be captured (such as, for example, Author, Data, and Title) and also frequently enforces, or at least suggests, rules for what can go into each data field.  Both the traditional library catalog and a spreadsheet of items, which assignes values to fields in typed columns (for example, an author column or a data column) are structured data.  

# III.  Defining Structures for Structured Data: Frameworks, Formats, and Standards

The overall value of structured data is directly proportional to the extent to which both individual and communties of catalogers apply the same structure to data descriptions of multiple artifacts.  If catalogers used a different structure each time they encountered a new object, the data they produce, would, from a practical standpoint, cease be structured, because the type of data being collected and the mode of representing that data would varry too widely to allow analysis across records.  

To address this problem, cultural heritage institutions typically adhere strict, common sets of rules that describe appropripriate data structures. I common parlance (as opposed to computing parlance), these rules are typically known as *Frameworks*, *Formats*, or *Standards*. This has te advantage of normalizing data across many collections.  However, it has the concommitant effect of flatting descriptions.  A frquent problem encountered by cataloguers is that defined structures may provide no way of representing something of interest in or on an artifict being described.

The above problem as has led, over time, to evolution and adoption of multiple frameworks for instantiating structured data, each of which speaks to a particular type of object (book, work of art, etc.), the type of institution creating the data (library, museum, etc.), or the perceived use of the data (contained catalog, networked data, etc.). # IV.  Metadata Frameworks Commonly Implemented by Libraries

Today, Library metadata is typically presented in one of the following three primary Framewors:


1.   [MARC (MAchine-Readable Cataloguing) Standard:](https://www.loc.gov/marc/)  MARC was developed by the LIbrary Congress in the 1960's in order to provide a standard structure for libaray catalogues in order to facilitate machine storage, transfer, and retreival of library metadata.  The original set of rules was based on the data feilds found on the typcial printed library card catalog, and grew over time into a complex, rich, standard for representing structured metadata about books.  MARC records are tupically stored on the computer as non-human-readable, binary files.  However, they can also be rendered as human readable XML files. MARC still serves as the primary data format for Worldcat and most library cataloguing systems.
2.   [BIBRRAME:](https://www.loc.gov/bibframe/)  In response to the ever-growing depth of information available over the World Wide Web, several years ago the Library of Congress began devlopig this new Framework for structured bibliographic data.  It is specifically designed to function in the networked world of [Linked Data](https://en.wikipedia.org/wiki/Linked_data), as opposed to the older, centeral repository world of the contained library catalog.  While libraries have devoted a tremendous amount of time, energy, and money to transitioning to Linked Data, as of today MARC remains the mainstay of library structured data representation.  You will learn more about the specifics of linked data in a future module of this course.
3.   [Dublin Core Metadata Initiative (DCMI):](https://dublincore.org/) The DCMI evolved in the early days of the World Wide Web as a means of inserting metadata into the code behind webpages.   It presents a minimal (sparse, in computing terminology) set of fields and rules for desscribing objects.  Many Libraries and Museums use DCMI standards to deliver metadata about digital objecs presented on the Web.
